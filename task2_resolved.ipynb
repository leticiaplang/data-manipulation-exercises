## Dataset
* US crimes informations between 1960 and 2014

### Task 01
* Preparing notebook

```python
# Imports
import pandas as pd
```

```python
# Accessing the document
crime = pd.read_csv('US_Crime_Rates_1960_2014.csv', sep=',')
```

```python
# Checking df
crime.head()
```

### Task 02
* Show teh columns type

```python
crime.info()
```

### Task 03
* Convert Year to datetime

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
crime.head()
```

### Task 04
* Set the Year as DF index

```python
crime = crime.set_index(crime['Year'])
crime.head()
```

### Task 05
* Remove the column Total

```python
crime = crime.drop(columns='Total')
crime.head()
```

### Task 06
* Get the total number of vehicle theft at 1978

```python
crime.drop(columns='Year', inplace=True)
```

```python
crime[crime.index == '1978-01-01'][['Vehicle_Theft']]
```

### Task 07
* Get the year with the highest murder total values

```python
crime[['Murder']].groupby(crime.index).agg(sum).sort_values('Murder', ascending=False).head()
```

### Task 08
* Get the murder number at the lowest vehicle theft year 

```python
crime[['Vehicle_Theft', 'Murder']].groupby(crime.index).agg('sum').sort_values('Vehicle_Theft').head()
```
